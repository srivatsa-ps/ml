# Step:1 Import Necessary Libraries

---



```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, accuracy_score, precision_score, recall_score
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb 
```

# Step 2: Load dataset

```python
CHP_df = pd.read_csv("sample_data/california_housing_test.csv") 
CHP_df.head()
```

# Step 3: Feature Extraction - split dataset in features and target variable 

```python

CHP_cols = ['longitude','latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income'] 
X = CHP_df[CHP_cols]# Features 
y = CHP_df.median_house_value# Target variable
feature_names = X
labels = y.unique()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0) 
```

# Step 4: Modelling and prediction

```python
# instantiate the model 

clf = DecisionTreeClassifier(max_depth =3, random_state = 42)

# fit the model with data 

clf.fit(X_train, y_train)

# predicting 
pred_decision_tree = clf.predict(X_test)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(30,10), facecolor ='k')

a = tree.plot_tree(clf,

                   rounded = True,

                   filled = True,

                   fontsize=14)

plt.show()
```

# Step 5: Performance of the model

```python
print("Precision Score : ",precision_score(y_test, pred_decision_tree , average='micro'))
print("Recall Score : ",recall_score(y_test, pred_decision_tree , average='micro'))
print("Accuracy Score : ",accuracy_score(y_test, pred_decision_tree ))

```

Show confusion matrix.

```python
cnf_matrix = confusion_matrix(y_test, pred_decision_tree )
print(cnf_matrix)
```
